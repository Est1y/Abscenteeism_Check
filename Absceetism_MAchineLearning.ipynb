```python
data_processed = data_processed.drop('Unnamed: 0', axis=1)
```

# Perform a logistic regression to predict Abscenteeism 

# Create targets

```python
# Using the Abscenteeism time in hour, classify the hour into excessive hour or normal
data_processed['Absenteeism Time in Hours'].median()
```

```python
targets = np.where(data_processed['Absenteeism Time in Hours'] > data_processed['Absenteeism Time in Hours'].median(),1, 0)
```

```python
targets
```

```python
data_processed['Excessive Abscenteeism'] = targets
```

```python
data_processed.head()
```

```python
targets.sum() / targets.shape[0]
```

```python
data_with_targets = data_processed.drop('Absenteeism Time in Hours', axis=1)
```

```python
data_with_targets.head()
```

```python
data_processed is data_with_targets
```

# Select Regression input

```python
data_with_targets.iloc[ :, : -1]
```

```python
Unscaled_input = data_with_targets.iloc[ :, : -1]
```

# Standardize data

```python
from sklearn.preprocessing import StandardScaler

abscentism_scalar = StandardScaler()
# Create Scalar object
```

```python
abscentism_scalar.fit(Unscaled_input) # Scaling mechanism
```

```python
Scaled_input = abscentism_scalar.transform(Unscaled_input) # Apply the scaling mechanism
```

```python
Scaled_input
```

```python
Scaled_input.shape
```

# Split the data into train & test and shuffle

```python
from sklearn.model_selection import train_test_split
```

# SPlit

```python
train_test_split(Scaled_input, targets)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(Scaled_input, targets, train_size=0.8, random_state= 20)
```

```python
X_train.shape, Y_train.shape
```

```python
X_test.shape, Y_test.shape
```

# Modelling

# Logistic regression with Sklearn

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

# Train the model

```python
reg = LogisticRegression()
reg.fit(X_train, Y_train)
```

```python
reg.score(X_train, Y_train)
```

Manaully check accuracy

```python
model_output = reg.predict(X_train)
model_output
```
